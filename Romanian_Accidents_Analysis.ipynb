12. `Latime banda`: sunt prezentate 4 categorii: 

    - *Intre 3.25 m si 3.75 m* 

    - *Intre 2.75 m si 3.25 m*

    - *Sub 2.75 m* 

13. `Drum compozitie`: Indica materialul din care este facut drumul: 

    - *Asfalt* 

    - *Piatra* 

    - *Pamant* 

    - *Beton* 

    - *Piatra cubica* 

14. `Drum stare`: starea drumului pe care a avut loc accidentul. 

15. `Drum aderenta`: aderenta drumului pe care a avut loc accidentul. Cuprinde 8 valori: 

    - *Uscat* 

    - *Umed* 

    - *Alunecos* 

    - *Mazga* 

    - *Gheata* 

    - *Zapada* 

    - *Polei* 

    - *Altele* 

16. `Cauza principala`: reprezinta cauza principala a accidentului. In setul de date sunt prezentate 37 de cauze principale. Cateva dintre acestea sunt: 

  - *Conducere fara permis* 

  - *Viteza neadaptata la conditiile de drum* 

  - *Abateri biciclisti* 

  - *Circulatie pe sens opus* 

  - *Adormire la volan* 

17. `Numar vehicule`: Numarul de vehicule implicate in accidentul rutier 

18. `Morti`: Numarul de persoane decedate in urma accidentului rutier 

19. `Raniti grav`: Numarul de persoane ranite grav in accidentul rutier 

20. `Vreme`: Conditiile meteo de la momentul producerii accidentului 

21. `Luminozitate`: Luminozitatea atmosferei 

22. `Lat`: Latitudinea locului producerii accidentului

23. `Long`: Longitunea locului producerii accidentului

24. `Configuratie`: Configuratia drumului: ex: curba L, in cruce, T

25. `Inclinatie`: Daca drumul a fost panda sau drum drept

26. `Intersectie cu`: Daca accidentul s-a produs in intersectie, e precizat numele strazii cu care strada se intersecteaza

27. `Drum restrictii`: Limitare de viteza, nerestrictionat, depasire interzisa, oprire interzisa etc

28. `Cauza secundara`: Detalii la cauza secundara ca "conducere sub influenta alcoolului", "circulatie pe sens opus"

29. `Mod producere`: Modul in care s-a produs accidentul: "cadere din vehicul", "lovire pieton"

30. `Numar vehicule`: Numarul de vehicule implicate

31. `Raniti usor`: Numarul de raniti usor

32. `Categorie banda`: Normala

33. `Marcaj drum`: ex: "separarea benzilor", "inexistent"

# 7. 🗃️ Functii folosite in analiza:

````python
def parse_lat_long(x):
  '''
  Parseaza coordonatele geografice in formatul corect.

  Exemplu: 
  In setul de date, coordonata pentru latitudine '023E13.3213' (tip de date: str)
  se va transforma in 23.2220 (tip de date: float).
  '''
  
  try:
    return parse(x)
  except:
    return np.NAN
````

````python
def create_dataframe(df, col):
  '''
  Creeaza un dataframe obtinut prin gruparea numarului de accidente din 
  dataframe-ul 'df' in functie de variabila 'col'.
  Acest dataframe nou creat va fi folosit in functiile de plotare.  
  '''
  
  temp = pd.DataFrame(df[col].value_counts().reset_index())
  temp.columns = [col, 'nr_accidente']
  temp['procent'] = round(temp['nr_accidente'] / df.shape[0] * 100, 2)

  return temp
````

````python
def plot_cat_vars(x_column, data, title, xlabel, total, kind,
                     fig_size=(14,8), y_column='nr_accidente', 
                     ylabel='Numarul accidentelor', legend=False, order=True):
  '''
  Ploteaza graficele (countplots sau barplots - dupa caz) pentru variabile categorice.
  Pentru fiecare variabila categorica din setul de date, vom plota numarul de accidente.
  '''

  fig, ax = plt.subplots(figsize=fig_size)
  ax.tick_params(axis='both', labelsize=14)

  if (kind=='bar'):
    # plot barplot
    if (legend):
      ax = sns.barplot(x=x_column,
                       y=y_column, 
                       data=data,
                       palette=colors,
                       hue=legend,
                       dodge=False)
    else:
      ax = sns.barplot(x=x_column, 
                       y=y_column, 
                       data=data, 
                       palette=colors)
  else:
    # plot countplot
    if(order):
     ax = sns.countplot(x=data[x_column], 
                        data=data,
                        palette=colors, 
                        order=data[x_column].value_counts().index)
    else:
     ax = sns.countplot(x=data[x_column],
                        data=data,
                        palette=colors)


  for p in ax.patches:
      x=p.get_bbox().get_points()[:,0]
      y=p.get_bbox().get_points()[1,1]
      ax.annotate('{:.2f}%'.format((p.get_height()/total)*100), (x.mean(), y), 
                  ha='center',
                  va='bottom', 
                  size=12, 
                  fontsize=15)

  plt.title(title, 
            fontdict={'fontsize' : 23}, 
            pad=10)
  plt.xlabel(xlabel,
             fontsize=15)
  plt.ylabel(ylabel
             ,fontsize=15)

  # fig.savefig('img{}.png'.format(no_figure), dpi=300)
  # no_figure += 1
  
  sns.despine()

  plt.show() 
````

## Numarul de accidente care au inregistrat persoane decedate:

````python
# studierea mai detaliata a accidentelor in care au murit persoane la momentul accidentului

# numarul total de accidente in care a murit cel putin o persoana
no_acc_total_deaths = traffic_accidents[traffic_accidents['morti'] > 0].shape[0]

# numarul de accidente in care a murit o singura persoana
no_acc_deaths1 = traffic_accidents[traffic_accidents['morti'] == 1].shape[0]

# numarul de accidente in care au murit cel putin doua persoane
no_acc_deaths2 = traffic_accidents[traffic_accidents['morti'] == 2].shape[0]

# numarul de accidente in care au murit cel putin trei persoane
no_acc_deaths_more3 = traffic_accidents[traffic_accidents['morti'] >= 3].shape[0]

deaths = {'Nr morti': ['> 1', '1', '2', '> 3'],
          'Nr accidente': [no_acc_total_deaths, 
                              no_acc_deaths1, 
                              no_acc_deaths2, 
                              no_acc_deaths_more3],
          'Procent': [1,
                      round(no_acc_deaths1 / no_acc_total_deaths * 100, 2), 
                      round(no_acc_deaths2 / no_acc_total_deaths * 100, 2),
                      round(no_acc_deaths_more3 / no_acc_total_deaths * 100, 2)
                      ]}

deaths_stats = pd.DataFrame(deaths)
deaths_stats
````

## Numarul de accidente din 2021 in care au fost ranite grav persoane:

````python
# studierea mai detaliata a accidentelor in care au fost ranite grav persoane
# ne referim la accidentele serioase, cele in care nu au murit oameni la locul
# accidentului, insa persoanele implicate in accident au fost ranite grav
# daca accidentul are persoane ranite grav si cel putin o persoana decedata = accident fatal

# numarul total de accidente serioase (cel putin un ranit grav dar 0 morti)
no_serious_acc = traffic_accidents[(traffic_accidents['raniti_grav'] > 0) &
                                  (traffic_accidents['morti'] == 0)].shape[0]

# numarul de accidente serioase cu o persoana ranita grav
no_serious_acc_1 = traffic_accidents[(traffic_accidents['raniti_grav'] == 1) &
                                    (traffic_accidents['morti'] == 0)].shape[0]

# numarul de accidente serioase cu doua persoane ranite grav                                    
no_serious_acc_2 = traffic_accidents[(traffic_accidents['raniti_grav'] == 2) & 
                                    (traffic_accidents['morti'] == 0)].shape[0]

# numarul de accidente serioase cu trei sau mai multe persoane ranite grav
no_serious_acc_more_3 = traffic_accidents[(traffic_accidents['raniti_grav'] >= 3) &
                                         (traffic_accidents['morti'] == 0)].shape[0]

                            
accidents_injured = {'Nr raniti grav': ['> 1', '1', '2', '> 3'],
          'Nr accidente': [no_serious_acc, 
                              no_serious_acc_1, 
                              no_serious_acc_2, 
                              no_serious_acc_more_3],
          'Procent': [1,
                      round(no_serious_acc_1 / no_serious_acc * 100, 2), 
                      round(no_serious_acc_2 / no_serious_acc * 100, 2),
                      round(no_serious_acc_more_3 / no_serious_acc * 100, 2)
                      ]}

accidents_injured_stats = pd.DataFrame(accidents_injured)
accidents_injured_stats
````
